```python
counts = df['Heart Disease'].value_counts()

plt.figure(figsize=(5,4))
plt.bar(counts.index, counts.values)
plt.xlabel("Heart Disease")
plt.ylabel("Number of Patients")
plt.title("Heart Disease Distribution")
plt.show()
```
